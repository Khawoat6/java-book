# การมีหลายรูปแบบ

ลองพิจารณาการขับรถ ถ้าเราขับรถเป็น เราสามารถขับรถอะไรก็ได้ (รถยนต์ทั่วไป) โดยไม่จำเป็นต้องรู้ว่าการทำงานภายในของรถแต่ละรุ่นแต่ละยี่ห้อแตกต่างกันอย่างไร เราสามารถขับรถโฟล์คเต่าได้เหมือน ๆ กับที่เราขับรถสปอร์ต เครื่องยนต์และระบบขับเคลื่อนของรถทั้ง 2 แบบนี้อาจจะแตกต่างกัน แต่ส่วนต่อประสานกับผู้ขับขี่ (interface) ก็เหมือน ๆ กัน คือมีส่วนสำหรับติดเครื่อง พวงมาลัย คันเร่ง เบรก กระปุกเกียร์ คลัตช์ ถ้าผู้ขับขี่เข้าใจส่วนต่อประสานเหล่านี้ก็สามารถขับขี่รถยนต์ได้ทุกแบบที่มีส่วนต่อประสานเหมือนกัน

ในรูปยกตัวอย่างคลาส `VolkswagenBeetle` และ `SportsCar` ซึ่งต่างก็สืบทอดมาจาก `Car` ผู้ขับขี่คือ `Driver` ผู้ขับขี่จำเป็นต้องรู้เพียงว่าการจะติดเครื่อง `Car` ทำได้โดยการเรียกเมทอด `start` โดยไม่จำเป็นต้องรู้ว่าขณะนั้น `Car` ที่กำลังใช้งานอยู่คือ `VolkswagenBeetle` หรือ `SportsCar` และไม่จำเป็นต้องรู้ว่า `start` ของ `VolkswagenBeetle` กับ `start` ของ `SportsCar` ทำงานต่างกันอย่างไร

คุณสมบัติที่ทำให้เราใช้อ็อบเจกต์ที่ต่างคลาสกันในรูปแบบเดียวกันได้แบบนี้เรียกว่าการมีหลายรูปแบบ (polymorphism)

คำว่า polymorphism เป็นคำประสมจากคำในภาษากรีก _polys_ แปลว่า "มาก" หรือ "หลาย" และ _morphē_ แปลว่า "รูปแบบ" หรือ "รูปร่าง" รวมกันจึงมีความหมายว่า "การมีหลายรูปแบบ" ในที่นี้เราจะใช้คำว่าการมีหลายรูปแบบและคำว่าโพลีมอร์ฟิซึมแทนความหมายเดียวกันตามความเหมาะสมในแต่ละบริบท

การมีหลายรูปแบบที่ปรากฏในการโปรแกรมมีหลายชนิด เช่น

- **การมีหลายรูปแบบแบบเฉพาะกิจ (ad hoc polymorphism)** การที่ฟังก์ชันหรือเมทอดสามารถแสดงพฤติกรรมที่แตกต่างกันโดยขึ้นอยู่กับรูปแบบของอาร์กิวเมนต์ที่ส่งผ่านเข้ามา การมีหลายรูปแบบแบบนี้จะขึ้นอยู่กับลักษณะเฉพาะเจาะจงของชนิดของอาร์กิวเมนต์แต่ละแบบไป จึงเรียกว่าเฉพาะกิจ (ad hoc) มักจะปรากฏในภาษาโปรแกรมในรูปของการโอเวอร์โหลดฟังก์ชันหรือเมทอด ซึ่งเป็นการประกาศฟังก์ชันหรือเมทอดหลายอันในชื่อเดียวกัน แต่มีชนิดของพารามิเตอร์ที่ต่างกัน

- **การมีหลายรูปแบบแบบอิงพารามิเตอร์ (parametric polymorphism)** การที่โค้ดไม่ได้ระบุชนิดของข้อมูลหรืออ็อบเจกต์ที่ใช้งานอย่างเฉพาะเจาะจง ทำให้สามารถใช้งานกับชนิดหรือคลาสใดก็ได้ตราบเท่าที่ชนิดหรือคลาสนั้นมีส่วนต่อประสาน (interface) ตามที่เรากำหนด เราพบการมีหลายรูปแบบแบบอิงพารามิเตอร์นี้ในรูปของเจเนอริกส์ (generics)

- **การกำหนดชนิดย่อย (subtyping)** การที่เราสามารถใช้ชื่อชื่อเดียวหรือตัวแปรตัวเดียวกันอ้างอิงข้อมูลหรืออ็อบเจกต์ที่ต่างคลาสกัน โดยที่คลาสที่ต่างกันเหล่านั้นมีซูเปอร์คลาสร่วมกัน กล่าวคือ เราสามารถอ้างอิงทั้งอ็อบเจกต์จากซูเปอร์คลาสและจากซับคลาสได้ด้วยตัวแปรตัวเดียวกัน คำว่าชนิดย่อยในที่นี้หมายถึงซับคลาส

- **การกำหนดชนิดแบบเป็ด (duck typing)** การที่เราสามารถใช้งานอ็อบเจกต์ใด ๆ ก็ได้โดยไม่ต้องสนใจชนิด ตราบเท่าที่ชนิดของอ็อบเจกต์นั้นมีคุณลักษณะหรือเมทอดที่เราต้องการใช้ การมีหลายรูปแบบลักษณะนี้ปรากฏในกลุ่มภาษาพลวัต เช่น Python และ Ruby เป็นต้น ที่เรียกว่าเป็ดมาจากคำกล่าวที่ว่า "ถ้ามันเดินเหมือนเป็ดและร้องเหมือนเป็ด มันจะต้องเป็นเป็ด" ซึ่งเปรียบเทียบกับการที่ไม่ต้องสนใจว่าจริง ๆ แล้วอ็อบเจกต์เป็นชนิดใด ขอแค่ให้มันมีคุณสมบัติตรงกับที่ต้องการก็ถือว่าใช้ได้

## การมีหลายรูปแบบในมุมมองของ OOP

ในบทนี้เราจะพิจารณาเฉพาะการกำหนดชนิดย่อย (subtyping) เนื่องจากในการโปรแกรมแบบ OOP เวลาที่เราพูดถึงการมีหลายรูปแบบเราจะหมายถึงการมีหลายรูปแบบแบบนี้ ภายใต้บริบทนี้เราสามารถนิยามคำว่าการมีหลายรูปแบบได้ว่า

> การมีหลายรูปแบบ (โพลีมอร์ฟิซึม) คือการที่เราสามารถส่งเมสเสจ (เรียกเมทอด) ไปยังอ็อบเจกต์ได้โดยไม่จำเป็นต้องรู้ชนิดที่แท้จริงของอ็อบเจกต์นั้น

แน่นอนว่าเราต้องรู้ว่าจะส่งเมสเสจอะไรได้บ้าง (มีเมทอดอะไรให้เรียกได้บ้าง) และอ็อบเจกต์ที่เราเรียกจะต้องมีเมทอดเหล่านั้น แต่เราไม่จำเป็นต้องรู้ว่าอ็อบเจกต์นั้นเป็นของคลาสอะไร

สิ่งที่จะทำให้รับรองได้ว่าอ็อบเจกต์ที่เราใช้งานมีเมทอดที่เราต้องการใช้คือการที่คลาสของอ็อบเจกต์นี้อยู่ในลำดับชั้นของคลาสที่มีเมทอดเหล่านี้ กล่าวคือ ถ้าคลาส A มีเมทอดที่เราต้องการใช้ อ็อบเจกต์ที่เราจะใช้งานต้องเป็นของคลาส A หรือซับคลาสของคลาส A

`Car`, `VolkswagenBeetle` และ `SportsCar` เป็นตัวอย่างหนึ่งของการมีหลายรูปแบบ เราสามารถเรียกเมทอด `start` ได้โดยไม่ต้องรู้ว่าอ็อบเจกต์ที่เราเรียกมาจากคลาส `Car`, `VolkswagenBeetle` หรือ `SportsCar` เรานับทั้ง 3 อย่างนี้เป็น `Car` เหมือนกันหมด และมี `start` เหมือนกันหมด

เราอาจจะมีเกมเกมหนึ่งที่มีสัตว์ 3 ประเภท ได้แก่ `Fish`, `Frog` และ `Bird` โดยเราสามารถสั่งให้สัตว์เหล่านี้เคลื่อนที่ได้ด้วยเมทอด `move`

สัตว์แต่ละชนิดมีลักษณะการเคลื่อนที่แตกต่างกัน ปลาอาจจะว่ายน้ำ กบกระโดด และนกบิน คลาสของสัตว์ทั้ง 3 ประเภทนี้อาจจะสืบทอดจากคลาส `Animal` ซึ่งกำหนดเมทอด `move` เอาไว้ ซึ่งโดยคุณสมบัติการสืบทอด คลาสทั้ง 3 คลาสที่สืบทอดมาก็จะรับเอาเมทอด `move` มาด้วย โดยอาจจะรับมาแล้วคงไว้หรือมีการโอเวอร์ไรด์ก็ได้ แต่ไม่ว่าจะกรณีใด ทั้ง `Fish`, `Frog` และ `Bird` ก็รับประกันได้ว่ามีเมทอด `move` ให้เรียกใช้

เราสามารถประกาศตัวแปรขึ้นมาให้มีชนิดเป็นตัวอ้างอิง (reference) ของคลาส `Animal` แต่ตัวแปรนี้จะสามารถใช้อ้างอิงอ็อบเจกต์ได้ทั้งที่มาจากคลาส `Animal`, `Fish`, `Frog` และ `Bird` เราสามารถสร้างอาร์เรย์ของ `Animal` โดยที่ในอาร์เรย์นี้เก็บอ็อบเจกต์ได้ทุกชนิดที่เป็น `Animal` และซับคลาสของมัน

คุณสมบัตินี้ทำให้เราสามารถพัฒนาระบบที่เพิ่มขยายในอนาคตได้ง่าย เราสามารถเพิ่มคลาสใหม่ได้โดยแทบจะไม่ต้องเปลี่ยนแปลงอะไรในโค้ดเดิมถ้าคลาสใหม่ที่เพิ่มเป็นส่วนหนึ่งของลำดับชั้นของคลาสที่โปรแกรมทำงานด้วยอยู่แล้ว อาจจะมีปรับแก้เฉพาะส่วนที่ต้องรู้จักคลาสใหม่โดยตรง เช่น ส่วนของการสร้างอ็อบเจกต์จากคลาสใหม่

พิจารณาตัวอย่างต่อไปนี้

````java
public class Animal {
    private String name;
    
    public Animal(String name) {
        this.name = name;
    }
    
    public String getName() {
        return name;
    }
    
    public void move() {
        System.out.println(getName() + ": How should I move?");
    }
}
````

เริ่มต้นจากคลาส `Animal` ซึ่งเรากำหนดให้มีเมทอด `getName` และ `move`

เราจะสร้างคลาส `Fish` ซึ่งสืบทอดจาก `Animal`

````java
public class Fish extends Animal {
    public Fish(String name) {
        super(name);
    }
    
    @Override
    public void move() {
        System.out.println(getName() + ": I'm swimming.");
    }
}
````

คลาส `Fish` โอเวอร์ไรด์เมทอด `move` เนื่องจากปลามีรูปแบบการเคลื่อนที่เฉพาะ ส่วนคอนสตรักเตอร์ก็ส่งต่อไปยังคลาส `Animal` ตรง ๆ

````java
public class Frog extends Animal {
    public Frog(String name) {
        super(name);
    }
    
    @Override
    public void move() {
        System.out.println(getName() + ": I'm jumping.");
    }
}
````

กบ (Frog) ก็เคลื่อนที่โดยการกระโดด จึงมีการโอเวอร์ไรด์ลักษณะเดียวกับปลา เช่นเดียวกับนก (Bird)

````java
public class Bird extends Animal {
    public Bird(String name) {
        super(name);
    }
    
    @Override
    public void move() {
        System.out.println(getName() + ": I'm flying.");
    }
}
````

เราจะทดสอบคลาสทั้งหมดนี้ด้วยโค้ดทดสอบต่อไปนี้

````java
System.out.println("* Initialization with the matching class:");
Animal nobody = new Animal("Nobody");
Fish nemo = new Fish("Nemo");
Frog kermit = new Frog("Kermit");
Bird tweety = new Bird("Tweety");
nobody.move();
nemo.move();
kermit.move();
tweety.move();

System.out.println("* Initialization with a subclass:");
Animal someone = new Bird("Polly");
someone.move();

System.out.println("* Upcasting assignment:");
Animal anyone;
anyone = kermit;
anyone.move();
anyone = nemo;
anyone.move();
````

ตัวแปร `nobody`, `nemo`, `kermit` และ `tweety` ถูกประกาศและสร้างอ็อบเจกต์ให้ด้วยคลาสที่ตรงกับชนิดของตัวแปรที่ประกาศ ผลการทำงานก็ตรงที่คาดเอาไว้ คือเมทอด `move` ที่ถูกเรียกเป็นไปตามคลาสของอ็อบเจกต์ ถึงตรงนี้เรายังไม่ได้เห็นคุณสมบัติการมีหลายรูปแบบ

จุดที่น่าสนใจอยู่ที่ตัวแปร `someone` ซึ่งเราประกาศให้เป็นคลาส `Animal` แต่สร้างอ็อบเจกต์ให้ด้วยคลาส `Bird` และตัวแปร `anyone` ซึ่งเราประกาศให้เป็นคลาส `Animal` และกำหนดค่าให้จากตัวแปร `kermit` ซึ่งเป็นอ็อบเจกต์ของคลาส `Frog` แล้วเปลี่ยนไปกำหนดค่าให้จากตัวแปร `nemo` ซึ่งมีคลาสเป็น `Fish` ในภายหลัง

ทั้ง `someone` และ `anyone` มีชนิดของตัวแปรตามที่ประกาศเป็นตัวอ้างอิงของคลาส `Animal` แต่อ็อบเจกต์ที่อ้างอิงเป็นคลาส `Bird`, `Frog` และ `Fish` ซึ่ง Java อนุญาตให้ทำได้ และเมื่อเรียกเมทอด `move` ตัวเมทอดที่ถูกเรียกก็จะเป็นไปตามชนิดของอ็อบเจกต์ ไม่ใช่ชนิดของตัวแปรหรือตัวอ้างอิง

การอนุญาตให้กำหนดอ็อบเจกต์ของซับคลาสให้กับตัวแปรที่ประกาศเป็นซูเปอร์คลาสได้ และการเลือกเมทอดที่จะทำงานตามชนิดของอ็อบเจกต์ที่อ้างอิงอยู่ในขณะนั้น เป็นคุณสมบัติของการมีหลายรูปแบบ

การเลือกเมทอดที่จะทำงานตามชนิดของอ็อบเจกต์ที่อ้างอิงอยู่ในแต่ละขณะเป็นกระบวนการที่เกิดขึ้นตอนโปรแกรมกำลังทำงาน ไม่ได้เลือกตั้งแต่ตอนคอมไพล์ เพราะตอนคอมไพล์เราไม่สามารถแน่ใจได้ว่าในขณะใดขณะหนึ่งตัวแปรจะอ้างอิงถึงอ็อบเจกต์ใด กระบวนการเลือกเมทอดขณะกำลังรันนี้เราเรียกว่า การจ่ายงานแบบพลวัต (dynamic dispatch)

### การสืบทอดและการมีหลายรูปแบบ

การมีหลายรูปแบบของ Java มีความสัมพันธ์อย่างเหนียวแน่นกับการสืบทอด เราอนุญาตให้มีลักษณะหลายรูปแบบกับกลุ่มคลาสที่มีความสัมพันธ์กันในลำดับชั้น เราอนุญาตให้ใช้อ็อบเจกต์ของซับคลาสได้ในทุกที่ที่ต้องการซูเปอร์คลาส ดังนั้นเราอาจกล่าวได้ว่าต้องมีการสืบทอดจึงจะมีการมีหลายรูปแบบได้

การสืบทอดเองโดยลำพังก็อาจจะไม่เกิดประโยชน์มากนัก หากมีเพียงการสืบทอดเอาคุณลักษณะและพฤติกรรมไป แต่ไม่อนุญาตให้มองซับคลาสเป็นซูเปอร์คลาสหรือให้ใช้อ็อบเจกต์จากซับคลาสในที่ที่ต้องการซูเปอร์คลาสได้ การสืบทอดก็จะเป็นเหมือนเพียงเครื่องมือสำหรับการใช้โค้ดซ้ำ เพียงเพื่ออำนวยความสะดวกเวลาที่ต้องการสร้างคลาสที่เพิ่มความสามารถเฉพาะจากคลาสเดิมโดยไม่ต้องเขียนโค้ดใหม่ทั้งหมด

การสืบทอดและการมีหลายรูปแบบจึงเป็นคุณสมบัติที่อยู่ร่วมกันและพึ่งพากันอย่างแยกไม่ได้

## คลาสและเมทอดนามธรรม

ตัวอย่าง `Animal` ข้างต้นมีประเด็นที่น่าสนใจอย่างหนึ่ง คลาส `Animal` โดยนัยแล้วไม่ได้เจาะจงถึงสัตว์ชนิดใดชนิดหนึ่ง แต่เป็นแนวคิด (concept) รวมของสัตว์นานาชนิด ถ้าเราถามว่ากบเคลื่อนที่อย่างไร เราตอบได้ว่าโดยการกระโดด หรือปลาเคลื่อนที่อย่างไร เราก็ตอบได้ว่าด้วยการว่ายน้ำ แต่ถ้าเราถามว่าสัตว์เคลื่อนที่อย่างไร เราคงตอบไม่ได้ เพราะคำว่าสัตว์เป็นคำครอบคลุมสัตว์ทุกชนิด แต่ละชนิดมีวิธีเคลื่อนที่ต่าง ๆ กันไป แม้กระทั่งถามว่าสัตว์มีรูปร่างหน้าตาอย่างไร เราก็คงตอบไม่ได้เช่นกัน ทั้งนี้เพราะสัตว์เป็นนามธรรมที่แทนกลุ่มของสิ่งมีชีวิตที่รวม ๆ กันเรียกว่าสัตว์ เราไม่สามารถระบุลักษณะหรือพฤติกรรมที่เจาะจงของสัตว์ได้

ในคลาส `Animal` ซึ่งเป็นตัวแทนของสัตว์ ถึงแม้จะมีเมทอด `move` แต่พฤติกรรมของ `move` สำหรับ `Animal` ก็ไม่สามารถนิยามได้ ในตัวอย่างที่ผ่านมา เราให้เมทอด `move` แสดงข้อความว่า "How should I move?" เพราะไม่รู้ว่าจะให้ทำอะไรดี ซึ่งในความเป็นจริงแล้วเราไม่ควรจะต้องกำหนดพฤติกรรมใด ๆ เลยด้วยซ้ำไป

Java อนุญาตให้เราสามารถประกาศเมทอดโดยไม่ต้องระบุพฤติกรรม (โค้ดของตัวเมทอด) ได้โดยการประกาศให้เมทอดนั้นเป็นนามธรรม ใช้คำว่า `abstract` หน้าชื่อเมทอด คลาสที่มีเมทอดที่เป็นนามธรรมจะถือเป็นคลาสนามธรรมไปด้วย และต้องระบุคำว่า `abstract` หน้าชื่อคลาสเช่นกัน

เรากำหนดคลาส `Animal` ใหม่ในแบบที่เป็นนามธรรมดังนี้

````java
public abstract class Animal {
    private String name;
    
    public Animal(String name) {
        this.name = name;
    }
    
    public String getName() {
        return name;
    }
    
    public abstract void move();
}
````

เมื่อพยายามรันโค้ดทดสอบนี้

````java
Animal nobody = new Animal("Nobody");
````

จะพบว่าไม่สามารถคอมไพล์ได้ โดยมีข้อความแสดงความผิดพลาด `Animal is abstract; cannot be instantiated` ซึ่งบ่งชี้ว่าคลาสที่เป็นนามธรรมไม่สามารถนำมาสร้างอ็อบเจกต์ได้

แต่ถึงกระนั้น เมื่อเรากำหนดคลาส `Fish` ให้สืบทอดจากคลาส `Animal` ที่เป็นนามธรรม

````java
public class Fish extends Animal {
    public Fish(String name) {
        super(name);
    }
    
    @Override
    public void move() {
        System.out.println(getName() + ": I'm swimming.");
    }
}
````

และทดสอบดังนี้

````java
Fish nemo = new Fish("Nemo");
nemo.move();
Animal someone = new Fish("Sushi");
someone.move();
someone = nemo;
someone.move();
````

เราจะพบว่าคลาส `Fish` ซึ่งสืบทอดมาจาก `Animal` ที่เป็นนามธรรมสามารถใช้สร้างอ็อบเจกต์ได้

คลาส `Animal` มีเมทอดที่เป็นนามธรรมคือ `move` ซึ่งสืบทอดมายัง `Fish` ด้วย แต่ `Fish` มีการโอเวอร์ไรด์เมทอดนี้ ทำให้ `move` ของ `Fish` ไม่เป็นนามธรรม ส่งผลให้ `Fish` ไม่ต้องเป็นคลาสนามธรรมไปด้วย และสามารถใช้สร้างอ็อบเจกต์ได้

คลาสที่ใช้สร้างอ็อบเจกต์ได้ต้องไม่เป็นนามธรรม ซึ่งหมายความว่าจะต้องไม่มีเมทอดนามธรรม หากสืบทอดเมทอดนามธรรมมาก็ต้องโอเวอร์ไรด์ให้หมดทุกเมทอดที่เป็นนามธรรม ไม่เช่นนั้นแล้วคลาสนั้นก็จะต้องเป็นคลาสนามธรรมไปด้วย ในทางกลับกัน เราสามารถระบุให้คลาสเป็นนามธรรมได้โดยไม่ต้องมีเมทอดนามธรรม และคลาสดังกล่าวจะไม่สามารถใช้สร้างอ็อบเจกต์ได้

ในตัวอย่างนี้เรายังเห็นการใช้ตัวแปร `someone` ซึ่งประกาศชนิดเป็นคลาส `Animal` ในการอ้างอิงอ็อบเจกต์จากคลาส `Fish` ด้วย โดยในส่วนแรก `Animal someone = new Fish("Sushi")` จะเห็นว่าเราสามารถสร้างอ็อบเจกต์ชนิด `Fish` แล้วกำหนดให้ `someone` ได้ เช่นเดียวกับ `someone = nemo` ซึ่งเป็นการกำหนดอ็อบเจกต์ชนิด `Fish` ซึ่งสร้างมาก่อนหน้านี้ให้กับ `someone`

เราอาจจะไม่สามารถสร้างอ็อบเจกต์จากคลาสนามธรรมได้ แต่เราสามารถประกาศตัวแปรให้มีชนิดเป็นคลาสนามธรรมได้ และตัวแปรนั้นสามารถอ้างอิงอ็อบเจกต์ต่าง ๆ ที่สร้างจากซับคลาสที่ไม่เป็นนามธรรมของคลาสนามธรรมนั้นได้

การที่คลาส `Animal` จำเป็นต้องประกาศเมทอด `move` ทั้ง ๆ ที่ตัวเองไม่สามารถจะ `move` ได้และต้องประกาศให้เป็นเมทอดนามธรรม ก็เพื่อเป็นสัญญาว่าซับคลาสทุก ๆ ซับคลาสของ `Animal` จะมีเมทอด `move` อยู่ให้เรียกใช้ได้ ถ้าคลาส `Animal` ไม่ประกาศเมทอด `move` เราจะไม่สามารถแน่ใจได้เลยว่าซับคลาสตัวไหนจะมีเมทอด `move` หรือไม่ และทำให้เราไม่สามารถใช้การมีหลายรูปแบบกับเมทอด `move` ได้

การประกาศเมทอดเอาไว้ในซูเปอร์คลาสเพื่อสร้างสัญญาว่าซับคลาสจะมีเมทอดอะไรให้เรียกใช้ได้บ้างนั้นเป็นหัวใจหลักอันหนึ่งของการออกแบบลำดับชั้นของคลาสเพื่อให้ทำงานตามที่เราต้องการ สัญญานี้เราเรียกว่าส่วนต่อประสาน (interface) ของคลาส เป็นข้อตกลงว่ากลุ่มคลาสเหล่านี้จะมีลักษณะภายนอกอย่างไร มีเมทอดอะไรบ้าง บางครั้งเราเรียกส่วนต่อประสานนี้ว่า สัญญา (contract) หรือข้อตกลงในการสื่อสาร เรียกอีกอย่างว่าโพรโทคอล (protocol)

เวลาที่เราเรียกเมทอด การพิจารณาว่ามีเมทอดที่เรียกหรือไม่จะพิจารณาจากคลาสของตัวอ้างอิงหรือตัวแปร แต่การเลือกว่าเมทอดที่ถูกเรียกจะเป็นตัวไหน (กรณีที่มีการโอเวอร์ไรด์) จะดูจากคลาสของอ็อบเจกต์ที่ถูกอ้างอิง

> **หมายเหตุ** ระวังการสับสนระหว่างคำว่า "ส่วนต่อประสาน" (interface) ในที่นี้ซึ่งเราหมายถึงสัญญาหรือข้อตกลงว่าคลาสจะมีลักษณะภายนอก (เมทอดที่เรียกใช้ได้) อย่างไรบ้าง กับคำว่า "อินเทอร์เฟซ" (interface) ที่เป็นคำเฉพาะที่หมายถึงหน่วยของโปรแกรมอีกแบบหนึ่งที่คล้ายกับคลาสซึ่งเราจะพูดถึงอีกครั้ง เราจะแยกความแตกต่างโดยการใช้คำว่า "ส่วนต่อประสาน" กับคำว่า "อินเทอร์เฟซ" เพื่อแทนความหมายที่แตกต่างกัน ทั้งสองคำนี้มาจากคำภาษาอังกฤษคำเดียวกัน แต่ความหมายต่างกันเมื่อใช้ในบริบทที่ต่างกัน

## ความสัมพันธ์แบบ is-a และ has-a

เราเคยพูดถึงความสัมพันธ์ในลำดับชั้นของคลาสไปก่อนหน้านี้แล้ว คลาสที่เป็นซับคลาสจะถือว่ามีความสัมพันธ์แบบ is-a กับซูเปอร์คลาส กล่าวคือ ซับคลาส is-a ซูเปอร์คลาส หรือ ซับคลาส "คือ" ซูเปอร์คลาส จากตัวอย่าง `Animal` เราพูดได้ว่า `Fish` is-a `Animal` หรือ `Fish` คือ `Animal` ซึ่งเมื่อเทียบกับความหมายจริง ๆ ปลาคือสัตว์ ก็สอดคล้องกันดี

ซูเปอร์คลาสจะมีลักษณะกว้าง ๆ (general) แต่ซับคลาสจะเป็นลักษณะที่เฉพาะลงไป (specific) ของซูเปอร์คลาส ดังนั้นเราจึงเห็นได้ว่าซับคลาสถือเป็นเซ็ตย่อยของซูเปอร์คลาสด้วย เราพูดได้ว่าซับคลาสคือซูเปอร์คลาสแบบหนึ่ง

ความสัมพันธ์ลักษณะที่ทำให้เราเห็นได้ว่าทำไมเราจึงสามารถใช้อ็อบเจกต์ของซับคลาสแทนที่ในส่วนที่ต้องการซูเปอร์คลาสได้เสมอ

ในทางกลับกัน เราไม่สามารถพูดได้ว่าซูเปอร์คลาสคือซับคลาส เราไม่สามารถพูดได้ว่า `Animal` is-a `Fish` หรือสัตว์คือปลา เพราะ `Animal` หรือซูเปอร์คลาสโดยทั่วไปมีความหมายที่กว้างกว่า

เราจึงไม่สามารถใช้อ็อบเจกต์ของซูเปอร์คลาสแทนที่ในที่ที่ต้องซับคลาสได้

ในทางเทคนิค ซับคลาสเองก็มีเมทอดของซูเปอร์คลาสทุกเมทอดเพราะสืบทอดมา จึงสามารถใช้แทนที่ได้ แต่ซูเปอร์คลาสไม่จำเป็นต้องมีเมทอดที่ซับคลาสมีทุกเมทอด ซูเปอร์คลาสจึงไม่สามารถใช้แทนที่ซับคลาสได้

is-a เป็นความสัมพันธ์แบบถ่ายทอด หมายความว่าถ้า `Penguin` is-a `Bird` และ `Bird` is-a `Animal` จะได้ว่า `Penguin` is-a `Animal` ด้วย ซึ่งในที่นี้ก็คือ `Animal` เป็นซูเปอร์คลาสของ `Penguin` ด้วย แต่เป็นซูเปอร์คลาสโดยอ้อม

นอกจากความสัมพันธ์แบบ is-a แล้ว ความสัมพันธ์อีกรูปแบบหนึ่งที่เรามักพิจารณาเมื่อออกแบบคลาสคือความสัมพันธ์แบบ has-a

ความสัมพันธ์แบบ has-a คือการที่คลาสหนึ่งมีอ็อบเจกต์ของอีกคลาสหนึ่งเป็นส่วนประกอบหนึ่งของมัน

พิจารณากลุ่มคลาสต่อไปนี้

````java
public class Wheel {
    // ... code omitted
}

public class Engine {
    // ... code omitted
}

public class DieselEngine extends Engine {
    // ... code omitted
}

public class Car {
    private Wheel[] wheels;
    private Engine engine;
    
    public Car(Wheel[] wheels, Engine engine) {
        wheels = wheels;
        engine = new Engine();
    }
    
    // ... code omitted
    
}

public class PickupTruck extends Car {
    public PickupTruck() {
        super(new Wheel[4], new DieselEngine());
    }
    
    // ... code omitted
    
}
````

ตัวอย่างนี้เรามองได้ว่า `DieselEngine` is-a `Engine` และ `PickupTruck` is-a `Car` ตามลักษณะการสืบทอด

ในคลาส `Car` เองมีคุณลักษณะ `wheels` ที่เป็นอาร์เรย์ของอ็อบเจกต์ของคลาส `Wheel` และ `engine` ที่เป็นอ็อบเจกต์ของคลาส `Engine` เรียกได้ว่า `Car` มี `Wheel` และ `Engine` เป็นส่วนประกอบ เรากล่าวได้ว่า `Car` has-a `Wheel` และ `Car` has-a `Engine` ด้วย ทั้งนี้คำว่า has-a เป็นชื่อของความสัมพันธ์ ไม่ได้บ่งบอกว่าต้องมีแค่อ็อบเจกต์เดียวจากที่คำนำหน้านาม a ใน has-a อาจชวนให้คิดอย่างนั้นได้ ในตัวอย่างนี้ `Car` มี `Wheel` ได้หลายอ็อบเจกต์โดยเก็บเป็นอาร์เรย์ของ `Wheel`

ในบางที่อาจจะแยกความสัมพันธ์แบบเป็นส่วนหนึ่งออกละเอียดขึ้นเป็น has-a, part-of และ member-of แล้วแต่ลักษณะของการจัดเก็บและความเป็นเจ้าของด้วย แต่ในที่นี้เราจะเรียกความสัมพันธ์ลักษณะนี้รวม ๆ ว่า has-a ทั้งหมด

## กรณีศึกษา: ระบบบัญชีเงินเดือน

ตัวอย่างต่อไปนี้เป็นการใช้คุณสมบัติการมีหลายรูปแบบเพื่อสร้างกลุ่มคลาสสำหรับแทนพนักงานแต่ละประเภท ได้แก่ พนักงานเงินเดือน (`SalariedEmployee`) พนักงานจ้างรายชั่วโมง (`HourlyEmployee`) พนักงานคอมมิชชัน (`CommissionEmployee`) และพนักงานเงินเดือนและคอมมิชชัน (`BasePlusCommissionEmployee`) โดยทั้งหมดนี้สืบทอด (ทั้งโดยตรงและโดยอ้อม) จากคลาส `Employee` ซึ่งเป็นคลาสนามธรรม

````java
public abstract class Employee {
    private final String firstName;
    private final String lastName;
    private final String nationalIdNumber;
    
    public Employee(String firstName, String lastName, String nationalIdNumber) {
        this.firstName = firstName;
        this.lastName = lastName;
        this.nationalIdNumber = nationalIdNumber;
    }
    
    public String getFirstName() {
        return firstName;
    }
    
    public String getLastName() {
        return lastName;
    }
    
    public String getNationalIdNumber() {
        return nationalIdNumber;
    }
    
    @Override
    public String toString() {
        return String.format("%s %s%nNational ID Number: %s",
                getFirstName(), getLastName(), getNationalIdNumber());
    }
    
    public abstract double getEarnings();
}
````

คลาส `Employee` โอเวอร์ไรด์เมทอด `toString` เพื่อกำหนดรูปแบบการแสดงข้อมูลพนักงานเบื้องต้น และกำหนดเมทอด `getEarnings` ไว้เป็นเมทอดนามธรรม เพื่อสร้างข้อตกลงว่าซับคลาสของ `Employee` ทุกซับคลาสจะต้องมีเมทอด `getEarnings` ซับคลาสที่โอเวอร์ไรด์เมทอด `getEarnings` จะต้องคืนค่าเป็นรายรับที่พนักงานจะได้รับ

````java
public class SalariedEmployee extends Employee {
    private double salary;
    
    public SalariedEmployee(String firstName, String lastName, String nationalIdNumber, double salary) {
        super(firstName, lastName, nationalIdNumber);
        
        if (salary < 0.0) {
            throw new IllegalArgumentException("Salary must be >= 0.0");
        }
        
        this.salary = salary;
    }

    public void setSalary(double salary) {
        if (salary < 0.0) {
            throw new IllegalArgumentException("Salary must be >= 0.0");
        }
        
        this.salary = salary;        
    }
    
    public double getSalary() {
        return salary;
    }
    
    @Override
    public double getEarnings() {
        return getSalary();
    }
    
    @Override
    public String toString() {
        return String.format("Salaried Employee: %s%n%s: %,.2f", super.toString(), "Salary", getSalary());
    }
}
````

คลาส `SalariedEmployee` โอเวอร์ไรด์เมทอด `getEarnings` เพื่อคืนค่ารายรับเป็นเงินเดือน และโอเวอร์ไรด์เมทอด `toString` เพื่อเพิ่มเติมข้อมูลจาก `toString` เดิมของ `Employee`

````java
public class HourlyEmployee extends Employee {
    private double wage;
    private double hours;
    
    public HourlyEmployee(String firstName, String lastName, String nationalIdNumber, double wage, double hours) {
        super(firstName, lastName, nationalIdNumber);
        
        if (wage < 0.0) {
            throw new IllegalArgumentException("Wage must be >= 0.0");
        }
        
        if (hours < 0.0 || hours > 720.0) {
            throw new IllegalArgumentException("Hourse worked must be >= 0.0 and <= 720.0");
        }
        
        this.wage = wage;
        this.hours = hours;
    }
    
    public void setWage(double wage) {
        if (wage < 0.0) {
            throw new IllegalArgumentException("Wage must be >= 0.0");
        }
        
        this.wage = wage;
    }
    
    public double getWage() {
        return wage;
    }
    
    public void setHours(double hours) {
        if (hours < 0.0 || hours > 720.0) {
            throw new IllegalArgumentException("Hours worked must be >= 0.0 and <= 720.0");
        }
        
        this.hours = hours;
    }
    
    public double getHours() {
        return hours;
    }
    
    @Override
    public double getEarnings() {
        if (getHours() <= 160) { // no overtime
            return getWage() * getHours();
        } else {
            return 160 * getWage() + (getHours() - 160) * getWage() * 1.5;
        }
    }
    
    @Override
    public String toString() {
        return String.format("Hourly Employee: %s%n%s: %,.2f; %s: %,.2f",
                super.toString(), "Hourly Wage", getWage(), "Hours Worked", getHours());
    }
}
````

คลาส `HourlyEmployee` โอเวอร์ไรด์เมทอด `getEarnings` เพื่อคืนค่ารายรับซึ่งคำนวณจากจำนวนชั่วโมงที่ทำงานทั้งหมดและค่าจ้างต่อชั่วโมง โดยมีการคิดอัตราล่วงเวลาให้เมื่อเกิน 160 ชั่วโมง และโอเวอร์ไรด์เมทอด `toString` เพื่อเพิ่มเติมข้อมูลจาก `toString` เดิมของ `Employee`

````java
public class CommissionEmployee extends Employee {
    private double grossSales;
    private double commissionRate;
    
    public CommissionEmployee(String firstName, String lastName, String nationalIdNumber,
            double grossSales, double commissionRate) {
        super(firstName, lastName, nationalIdNumber);
        
        if (commissionRate <= 0.0 || commissionRate >= 1.0) {
            throw new IllegalArgumentException("Commission rate must be > 0.0 and < 1.0");
        }
        
        if (grossSales < 0.0) {
            throw new IllegalArgumentException("Gross sales must be >= 0.0");
        }
        
        this.grossSales = grossSales;
        this.commissionRate = commissionRate;
    }
    
    public void setGrossSales(double grossSales) {
        if (grossSales < 0.0) {
            throw new IllegalArgumentException("Gross sales must be >= 0.0");
        }
        
        this.grossSales = grossSales;
    }
    
    public double getGrossSales() {
        return grossSales;
    }
    
    public void setCommissionRate(double commissionRate) {
        if (commissionRate <= 0.0 || commissionRate >= 1.0) {
            throw new IllegalArgumentException("Commission rate must be > 0.0 and < 1.0");
        }
        
        this.commissionRate = commissionRate;        
    }
    
    public double getCommissionRate() {
        return commissionRate;
    }
    
    @Override
    public double getEarnings() {
        return getCommissionRate() * getGrossSales();
    }
    
    @Override
    public String toString() {
        return String.format("Commission Employee: %s%n%s: %,.2f; %s: %.2f",
                super.toString(), "Gross Sales", getGrossSales(),
                "Commission Rate", getCommissionRate());
    }
}
````

คลาส `CommissionEmployee` โอเวอร์ไรด์เมทอด `getEarnings` เพื่อคืนค่ารายรับเป็นค่าคอมมิชชัน โดยคิดจากอัตราค่าคอมมิชชันและยอดขาย และโอเวอร์ไรด์เมทอด `toString` เพื่อเพิ่มเติมข้อมูลจาก `toString` เดิมของ `Employee`

````java
public class BasePlusCommissionEmployee extends CommissionEmployee {
    private double baseSalary;
    
    public BasePlusCommissionEmployee(String firstName, String lastName, String nationalIdNumber,
            double grossSales, double commissionRate, double baseSalary) {
        super(firstName, lastName, nationalIdNumber, grossSales, commissionRate);
        
        if (baseSalary < 0.0) {
            throw new IllegalArgumentException("Base salary must be >= 0.0");
        }
        
        this.baseSalary = baseSalary;
    }
    
    public void setBaseSalary(double baseSalary) {
        if (baseSalary < 0.0) {
            throw new IllegalArgumentException("Base salary must be >= 0.0");
        }
        
        this.baseSalary = baseSalary;
    }
    
    public double getBaseSalary() {
        return baseSalary;
    }
    
    @Override
    public double getEarnings() {
        return super.getEarnings() + getBaseSalary();
    }
    
    @Override
    public String toString() {
        return String.format("Base-Salaried %s; %s: %,.2f", super.toString(), "Base Salary", getBaseSalary());
    }
}
````

คลาส `BasePlusCommissionEmployee` สืบทอดจากคลาส `CommissionEmployee` อีกที โดยโอเวอร์ไรด์เมทอด `getEarnings` เพื่อคืนค่ารายรับเป็นค่าคอมมิชชันซึ่งคำนวณโดยการเรียก `super.getEarnings()` ของ `CommissionEmployee` รวมกับเงินเดือนฐาน (base salary) และโอเวอร์ไรด์เมทอด `toString` เพื่อเพิ่มเติมข้อมูลจาก `toString` เดิมของ `CommissionEmployee`

เราทดสอบพฤติกรรมของการมีหลายรูปแบบในคลาสทดสอบ `PayrollSystemTest` โดยจะทดสอบ 3 รูปแบบดังนี้

1. ทดสอบสร้างและเรียกใช้อ็อบเจกต์จากคลาสพนักงานทั้ง 4 แบบโดยตรง โดยการกำหนดชนิดตัวแปรให้ตรงกับชนิดของอ็อบเจกต์ที่สร้างขึ้น
2. ทดสอบเรียกใช้อ็อบเจกต์ทั้ง 4 ชนิดผ่านตัวแปรชนิด `Employee` ตัวเดียวกัน โดยมีการทดสอบการตรวจสอบชนิดของอ็อบเจกต์และการแปลงชนิดลง (downcasting) ด้วย
3. ทดสอบการแสดงคลาสของอ็อบเจกต์แต่ละตัว

````java
public class PayrollSystemTest {
    public static void main(String[] args) {
        SalariedEmployee salariedEmployee =
                new SalariedEmployee("John", "Smith", "111-11-1111", 100_000.00);
        HourlyEmployee hourlyEmployee =
                new HourlyEmployee("Karen", "Price", "222-22-2222", 500.00, 160);
        CommissionEmployee commissionEmployee =
                new CommissionEmployee("Sue", "Jones", "333-33-3333", 1_200_000.00, .06);
        BasePlusCommissionEmployee basePlusCommissionEmployee =
                new BasePlusCommissionEmployee("Bob", "Lewis", "444-44-4444", 600_000.00, .04, 36_000.00);
        
        System.out.printf("Employees processed individually:%n%n");

        System.out.printf("%s%nEarned: %,.2f%n%n", salariedEmployee, salariedEmployee.getEarnings());
        System.out.printf("%s%nEarned: %,.2f%n%n", hourlyEmployee, hourlyEmployee.getEarnings());
        System.out.printf("%s%nEarned: %,.2f%n%n", commissionEmployee, commissionEmployee.getEarnings());
        System.out.printf("%s%nEarned: %,.2f%n%n", basePlusCommissionEmployee, basePlusCommissionEmployee.getEarnings());

        Employee[] employees = new Employee[4];
        
        employees[0] = salariedEmployee;
        employees[1] = hourlyEmployee;
        employees[2] = commissionEmployee;
        employees[3] = basePlusCommissionEmployee;

        System.out.printf("Employees processed polymorphically:%n%n");

        for (Employee currentEmployee : employees) {
            System.out.println(currentEmployee);
            
            // Raise salary for BasePlusCommissionEmployee
            if (currentEmployee instanceof BasePlusCommissionEmployee) {
                // Downcast Employee reference to BasePlusCommissionEmployee reference
                BasePlusCommissionEmployee employee =
                        (BasePlusCommissionEmployee)currentEmployee;
                
                employee.setBaseSalary(1.10 * employee.getBaseSalary());
                
                System.out.printf("New base salary with 10%% increase is: %,.2f%n", employee.getBaseSalary());
            }
            
            System.out.printf("Earned: %,.2f%n%n", currentEmployee.getEarnings());
        }
        
        // Show type name of each object in employees array
        for (int i = 0; i < employees.length; i++) {
            System.out.printf("Employee %d is a %s%n", i, employees[i].getClass().getName());
        }
    }
}
````

### การทดสอบสร้างและเรียกใช้อ็อบเจกต์จากคลาสพนักงานทั้ง 4 แบบโดยตรง

อ็อบเจกต์แต่ละตัวถูกสร้างและเรียกใช้ผ่านตัวแปรที่ตรงกับคลาสของอ็อบเจกต์นั้น ๆ ในส่วนนี้เราไม่ได้ใช้คุณสมบัติการมีหลายรูปแบบ

### การทดสอบเรียกใช้อ็อบเจกต์ทั้ง 4 ชนิดผ่านตัวแปรชนิด Employee ตัวเดียวกัน

เราใช้ลูป `for` ในการวนซ้ำเพื่อประมวลผลอ็อบเจกต์ทีละตัว เราสามารถทำแบบนี้ได้เพราะคุณสมบัติการมีหลายรูปแบบทำให้อ็อบเจกต์ทุกตัวสามารถอ้างอิงด้วยตัวแปร `currentEmployee` ซึ่งมีชนิดเป็น `Employee` ได้

เมทอด `getEarnings` ที่เราเรียกจะถูกเลือกตอนที่โปรแกรมทำงานโดยดูจากอ็อบเจกต์ปัจจุบันว่ามาจากคลาสใด ไม่ได้ดูจากชนิดของตัวแปร `currentEmployee`

สิ่งที่น่าสนใจอีกอย่างหนึ่งคือการปฏิบัติต่อคลาส `BasePlusCommissionEmployee` ที่แตกต่างจากคลาสอื่น

จุดแรกคือการใช้ตัวดำเนินการ `instanceof` ซึ่งใช้ทดสอบว่าอ็อบเจกต์เป็นอ็อบเจกต์จากคลาสที่ระบุหรือจากซับคลาสของคลาสที่ระบุหรือไม่

สมมติให้ `e` อ้างอิงอ็อบเจกต์ของคลาส `BasePlusCommissionEmployee` เราสรุปได้ว่า `e instanceof BasePlusCommissionEmployee` และ `e instanceof CommissionEmployee` จะให้ผลลัพธ์เป็น true และ `e instanceof HourlyEmployee` เป็น false

นอกจากนี้ยังมีตรงส่วนนี้

```java
BasePlusCommissionEmployee employee = (BasePlusCommissionEmployee)currentEmployee;

employee.setBaseSalary(1.10 * employee.getBaseSalary());
```

`(BasePlusCommissionEmployee)currentEmployee` คือการบังคับแปลงชนิดข้อมูลที่อ้างอิงโดย `currentEmployee` ให้เป็นชนิดคลาส `BasePlusCommissionEmployee` โดยที่ชนิดของตัวแปร `currentEmployee` คือ `Employee` และเมื่อแปลงแล้วให้กำหนดค่าให้กับตัวแปร `employee` ซึ่งมีชนิดเป็นคลาส `BasePlusCommissionEmployee`

การบังคับแปลงชนิดข้อมูลในลักษณะนี้ คือการแปลงจากตัวแปรที่มีชนิดเป็นซูเปอร์คลาสไปเป็นอ็อบเจกต์ที่มีชนิดเป็นซับคลาส เราเรียกว่าการแปลงชนิดลง (downcasting) ซึ่งจะสำเร็จได้ก็ต่อเมื่ออ็อบเจกต์ที่กำลังแปลงมีชนิดที่เข้ากันได้กับชนิดที่ต้องการแปลง (เป็นคลาสเดียวกันหรือเป็นซับคลาสของชนิดที่ต้องการแปลง) ถ้าชนิดของอ็อบเจกต์ไม่เข้ากับชนิดที่ต้องการแปลง จะเกิด `ClassCastException` ขึ้น

เหตุที่ต้องแปลงชนิดในที่นี้เนื่องจากเราต้องการเรียกเมทอด `getBaseSalary` ซึ่งมีเฉพาะในคลาส `BasePlusCommissionEmployee` การเรียกผ่านตัวแปร `currentEmployee` โดยตรงจะไม่สามารถทำได้ เนื่องจาก `currentEmployee` มีชนิดเป็นคลาส `Employee` ซึ่งไม่มีเมทอด `getBaseSalary` ขณะคอมไพล์ตัวคอมไพเลอร์จะไม่สามารถรู้ได้ว่าอ็อบเจกต์ที่ `currentEmployee` จะอ้างอิงขณะโปรแกรมทำงานจริงแต่ละช่วงจะมีเมทอด `getBaseSalary` หรือไม่ จึงต้องสั่งแปลงชนิดขณะโปรแกรมรันจริง ซึ่งจะมีการทดสอบว่าอ็อบเจกต์ขณะนั้นสามารถแปลงไปเป็น `BasePlusCommissionEmployee` ได้หรือไม่

ในที่นี้โปรแกรมนี้รับรองได้ว่าสามารถแปลงชนิดได้เสมอ เนื่องจากมีการทดสอบชนิดของอ็อบเจกต์ด้วย `instanceof` ก่อนแล้ว

การแปลงชนิดลงและการทดสอบชนิดเป็นสิ่งที่ควรหลีกเลี่ยงถ้าไม่จำเป็นจริง ๆ เนื่องจากจะทำให้โค้ดมีส่วนที่ทำงานแบบเฉพาะเจาะจงเกิดขึ้น มีลักษณะของการมีหลายรูปแบบลดลง การมีข้อยกเว้นจำนวนมากในโค้ดจะทำให้การดูแลรักษาและพัฒนาโค้ดต่อไปลำบากขึ้นได้

### การทดสอบการแสดงคลาสของอ็อบเจกต์แต่ละตัว

ส่วนสุดท้ายของการทดสอบเป็นการแสดงคลาสของอ็อบเจกต์แต่ละตัวโดยการเรียกเมทอด `getClass` ซึ่งจะได้ผลลัพธ์เป็นอ็อบเจกต์ชนิด `Class` ซึ่งเก็บข้อมูลของคลาสของอ็อบเจกต์ที่เราทดสอบ อ็อบเจกต์ `Class` จะมีเมทอด `getName` ให้เรียกต่อเพื่อให้ได้ชื่อของคลาสมา

## สรุปการกำหนดค่าระหว่างตัวแปรซับคลาสและซูเปอร์คลาส

การกำหนดค่าตัวอ้างอิงอ็อบเจกต์ให้กับตัวแปร เช่น ในรูปแบบนี้

```java
a = b;
```

จะสามารถทำได้ถ้า

1. `a` ประกาศเป็นคลาสเดียวกับอ็อบเจกต์ `b`
2. `a` ประกาศเป็นซูเปอร์คลาสของอ็อบเจกต์ `b` เพราะซับคลาสถือเป็นซูเปอร์คลาสด้วยความสัมพันธ์แบบ is-a

ทั้งนี้ ในกรณีที่สอง `a` จะสามารถใช้อ้างอิงได้เฉพาะสมาชิกของซูเปอร์คลาส ไม่สามารถเรียกเมทอดของซับคลาสได้

ในกรณีที่ `a` ประกาศเป็นซับคลาสของคลาสของอ็อบเจกต์ `b` เราจะไม่สามารถกำหนดค่า `b` ให้ `a` ได้นอกจากจะมีการแปลงชนิดลง

## ตัวกำหนด final กับคลาสและเมทอด

เราสามารถประกาศให้เมทอดหรือคลาสเป็น `final` ได้โดยมีความหมายดังนี้

เมทอดที่ถูกประกาศให้เป็น `final` จะไม่สามารถถูกโอเวอร์ไรด์ได้ในซับคลาส การเรียกเมทอดที่เป็น `final` คอมไพเลอร์จะรู้ได้ตั้งแต่ตอนคอมไพล์ว่าจะต้องเรียกเมทอดไหน เนื่องจากมีเพียงเมทอดเดียว ซับคลาสจะสืบทอดไปโดยไม่สามารถโอเวอร์ไรด์ได้ เมทอดที่ประกาศเป็น `private` หรือ `static` จะถือว่าเป็น `final` โดยปริยายเนื่องจากไม่สามารถโอเวอร์ไรด์ได้อยู่แล้ว

คลาสที่ถูกประกาศให้เป็น `final` จะไม่สามารถสืบทอดได้ เราจะไม่สามารถสร้างซับคลาสของคลาสที่กำหนดให้เป็น `final` ได้ และเมทอดทุกเมทอดของคลาสดังกล่าวจะถือเป็น `final` โดยปริยายเช่นกัน คลาส `String` เป็นตัวอย่างของคลาสที่เป็น `final`

## ปัญหาในการเรียกเมทอดจากคอนสตรักเตอร์

ในตัวอย่างกลุ่มคลาสพนักงานเราจะเห็นการทำงานที่ซ้ำซ้อนในคอนสตรักเตอร์และเมทอดกำหนดค่าในหลาย ๆ คลาส เช่น ใน `SalariedEmployee` ซึ่งในคอนสตรักเตอร์มีการตรวจสอบก่อนการกำหนดค่าให้กับ `salary` แต่ก็มีเมทอด `setSalary` ซึ่งตรวจสอบก่อนกำหนดค่าให้กับ `salary` เช่นกัน ในกรณีแบบนี้เราอาจจะอยากจะรวบโค้ดเพื่อไม่ให้มีโค้ดซ้ำซ้อนโดยการให้คอนสตรักเตอร์ไปเรียกเมทอด `setSalary` เลยและไม่ต้องตรวจสอบค่าเอง

Java อนุญาตให้คอนสตรักเตอร์เรียกเมทอดอื่น ๆ ได้ แต่ในทางปฏิบัติแล้วไม่ควรทำ เนื่องจากอาจเป็นไปได้ว่าในอนาคตอาจจะมีคลาสอื่นที่สืบทอดจาก `SalariedEmployee` แล้วไปโอเวอร์ไรด์เมทอด `setSalary` ซึ่งจะทำให้เกิดปัญหาเมื่อเราต้องการสร้างอ็อบเจกต์จากคลาสใหม่นี้

เมื่อเราสร้างอ็อบเจกต์จากคลาสใหม่ คอนสตรักเตอร์ของคลาสใหม่จะไปเรียกคอนสตรักเตอร์ของ `SalariedEmployee` ซึ่งถ้าในคอนสตรักเตอร์ของ `SalariedEmployee` ไปเรียกเมทอด `setSalary` เข้า ในกรณีนี้ เนื่องจากเรากำลังสร้างอ็อบเจกต์ของคลาสใหม่ `setSalary` ที่ถูกเรียกจะเป็นของคลาสใหม่ที่โอเวอร์ไรด์ `setSalary` เดิม และ `setSalary` ของคลาสใหม่นี้จะทำงานโดยที่คอนสตรักเตอร์ของคลาสใหม่ยังไม่ทันได้ทำงาน การกำหนดสถานะเริ่มต้นของอ็อบเจกต์ยังไม่เสร็จสิ้น ซึ่งอาจนำไปสู่การทำงานที่ผิดพลาดและยากที่จะหาต้นตอได้

ในกรณีที่เมทอดที่คอนสตรักเตอร์ต้องการเรียกเป็น `static` เราสามารถเรียกได้เลย เพราะเมทอด `static` ไม่สามารถโอเวอร์ไรด์ได้ จึงไม่เกิดปัญหาอย่างที่กล่าวมาข้างต้น

## อินเทอร์เฟซ

ถ้าบริษัทหนึ่งต้องการทำระบบบัญชีซึ่งคำนวณยอดค้างชำระ (payable) หลาย ๆ ประเภทในระบบเดียวกัน ในที่นี้สมมติว่ามีรายการที่ค้างชำระได้อยู่ 2 ประเภท ได้แก่ใบส่งของหรือใบกำกับสินค้า (invoice) และเงินค่าจ้างพนักงาน

ทั้ง 2 ประเภทนี้สิ่งที่เราต้องการคือการคำนวณยอดที่ต้องชำระ ถ้าเป็นใบกำกับสินค้า ยอดที่ต้องชำระก็คือราคารวมทั้งหมด (total price) ของสินค้าในรายการ แต่ถ้าเป็นพนักงาน ยอดที่ต้องชำระก็คือรายได้ (earnings) ของพนักงาน

ถ้าเราต้องการกำหนดเมทอดชื่อว่า `getPaymentAmount` โดยให้ทั้งคลาสที่แทนใบกำกับสินค้า (`Invoice`) และคลาสพนักงาน (`Employee`) มีเมทอดนี้ทั้งคู่ และให้เราสามารถจัดเก็บอ็อบเจกต์ของทั้งคลาส `Invoice` และคลาส `Employee` ไว้ร่วมกันในรายการบัญชีค้างชำระเดียวได้

จากตัวอย่างที่ผ่านมา เราสามารถรวมอ็อบเจกต์จากต่างคลาสกันไว้ด้วยกันเช่นในอาร์เรย์เดียวกันได้ก็ต่อเมื่อคลาสเหล่านั้นมีซูเปอร์คลาสร่วมกัน แต่ในกรณีนี้ `Invoice` กับ `Employee` โดยลักษณะแล้วไม่ควรจะมีซูเปอร์คลาสร่วมกัน เนื่องจากไม่มีความสัมพันธ์กันในเชิงความหมายที่เหมาะสม

Java มีเครื่องมืออีกอย่างที่ทำให้เราสามารถกำหนดเมทอดที่เราต้องการให้คลาสต่างชนิดกันมีขึ้นมาได้ เครื่องมือนั้นคืออินเทอร์เฟซ (interface)

อินเทอร์เฟซทำให้เราสามารถกำหนดเมทอดหรือชุดของเมทอดขึ้นมา และบังคับให้คลาสที่อิมพลีเมนต์ (implement) อินเทอร์เฟซนั้นต้องมีเมทอดเหล่านั้นได้ อินเทอร์เฟซจึงเป็นเหมือนเครื่องมือที่ทำให้เราสร้างข้อตกลงหรือสัญญาได้ว่ากลุ่มคลาสที่อิมพลีเมนต์อินเทอร์เฟซนี้จะต้องมีเมทอดเหล่านี้

นอกจากนี้ อินเทอร์เฟซยังทำให้เราสามารถอ้างอิงอ็อบเจกต์จากคลาสต่างชนิดกันที่อิมพลีเมนต์อินเทอร์เฟซเดียวกันได้ด้วยตัวแปรที่มีชนิดเป็นอินเทอร์เฟซนั้น คุณสมบัตินี้เหมือนกับการที่เราสามารถอ้างอิงอ็อบเจกต์ของคลาสต่าง ๆ ได้ด้วยตัวแปรของซูเปอร์คลาส

ข้อแตกต่างของอินเทอร์เฟซกับการใช้ซูเปอร์คลาสคือ

1. คลาสสามารถสืบทอดจากซูเปอร์คลาสได้ซูเปอร์คลาสเดียว (single inheritance) แต่สามารถอิมพลีเมนต์อินเทอร์เฟซที่แตกต่างกันได้ไม่จำกัด
2. เมทอดในอินเทอร์เฟซจะเป็นนามธรรมทั้งหมด ไม่มีส่วนของตัวโค้ด (ข้อจำกัดนี้ถูกยกเลิกไปใน Java SE 8 ซึ่งอนุญาตให้เมทอดในอินเทอร์เฟซมีโค้ดโดยปริยายได้ แต่เราจะยังไม่นำมาพิจารณาในตอนนี้)

การประกาศคลาสที่อิมพลีเมนต์อินเทอร์เฟซจะมีรูปแบบวากยสัมพันธ์ดังนี้

```
<class-modifiers> class SubclassName extends SuperclassName
        implements FirstInterface, SecondInterface, ... {
    ...
}
```

โดยหลักการแล้ว เราจะใช้ซูเปอร์คลาสเพื่อสร้างความสัมพันธ์แบบ is-a ระหว่างกลุ่มคลาส และให้การสืบทอดเป็นตัวส่งผ่านความสามารถไปยังซับคลาส แต่เราจะใช้อินเทอร์เฟซเพื่อกำหนดความสามารถหรือพฤติกรรมร่วมบางอย่างโดยที่กลุ่มคลาสที่จะมีความสามารถร่วมนั้นไม่จำเป็นต้องมีความเกี่ยวข้องกัน

พิจารณาการประกาศอินเทอร์เฟซ `Payable` ต่อไปนี้

````java
public interface Payable {
    double getPaymentAmount();
}
````

เราจะเห็นว่าอินเทอร์เฟซ `Payable` มีการประกาศเมทอด `getPaymentAmount` เอาไว้ เมทอดในอินเทอร์เฟซจะถือเป็นเมทอดนามธรรมและเป็นสาธารณะโดยปริยาย การประกาศอินเทอร์เฟซนี้เป็นการสร้างข้อตกลงหรือสัญญาว่าคลาสใดก็ตามที่อิมพลีเมนต์อินเทอร์เฟซนี้จะต้องมีเมทอดตามที่อินเทอร์เฟซประกาศไว้

คลาส `Invoice` อิมพลีเมนต์อินเทอร์เฟซ `Payable` ด้วยคำว่า `implements` และมีการโอเวอร์ไรด์เมทอด `getPaymentAmount` เพื่อคำนวณราคารวมเป็นยอดที่ต้องชำระ

````java
public class Invoice implements Payable {
    private final String partNumber;
    private final String partDescription;
    private int quantity;
    private double pricePerItem;
    
    public Invoice(String partNumber, String partDescription, int quantity, double pricePerItem) {
        if (quantity < 0) {
            throw new IllegalArgumentException("Quantity must be >= 0");
        }
        
        if (pricePerItem < 0.0) {
            throw new IllegalArgumentException("Price per item must be >= 0");            
        }
        
        this.partNumber = partNumber;
        this.partDescription = partDescription;
        this.quantity = quantity;
        this.pricePerItem = pricePerItem;
    }
    
    public String getPartNumber() {
        return partNumber;
    }
    
    public String getPartDescription() {
        return partDescription;
    }
    
    public void setQuantity(int quantity) {
        if (quantity < 0) {
            throw new IllegalArgumentException("Quantity must be >= 0");
        }
        
        this.quantity = quantity;
    }
    
    public int getQuantity() {
        return quantity;
    }
    
    public void setPricePerItem(double pricePerItem) {
        if (pricePerItem < 0.0) {
            throw new IllegalArgumentException("Price per item must be >= 0");            
        }
        
        this.pricePerItem = pricePerItem;
    }
    
    public double getPricePerItem() {
        return pricePerItem;
    }
    
    @Override
    public String toString() {
        return String.format("Invoice:%nPart Number: %s (%s)%nQuantity: %d%nPrice/Item: %,.2f",
                getPartNumber(), getPartDescription(), getQuantity(), getPricePerItem());
    }
    
    @Override
    public double getPaymentAmount() {
        return getQuantity() * getPricePerItem();
    }
}
````

คลาส `Employee` อิมพลีเมนต์อินเทอร์เฟซ `Payable` เช่นกัน แต่ไม่ได้โอเวอร์ไรด์เมทอด `getPaymentAmount` ซึ่งหมายความว่า `Employee` รับเอาเมทอด `getPaymentAmount` มาในรูปของเมทอดนามธรรม และทำให้คลาส `Employee` ต้องเป็นคลาสนามธรรม

ในตัวอย่างนี้สังเกตว่าจะไม่มีเมทอด `getEarnings` เหมือนในตัวอย่างก่อนหน้า ทั้งนี้เนื่องจากเราต้องการใช้ `getPaymentAmount` แทนที่ `getEarnings` ไปเลย

````java
public abstract class Employee implements Payable {
    private final String firstName;
    private final String lastName;
    private final String nationalIdNumber;
    
    public Employee(String firstName, String lastName, String nationalIdNumber) {
        this.firstName = firstName;
        this.lastName = lastName;
        this.nationalIdNumber = nationalIdNumber;
    }
    
    public String getFirstName() {
        return firstName;
    }
    
    public String getLastName() {
        return lastName;
    }
    
    public String getNationalIdNumber() {
        return nationalIdNumber;
    }
    
    @Override
    public String toString() {
        return String.format("%s %s%nNational ID Number: %s",
                getFirstName(), getLastName(), getNationalIdNumber());
    }
}
````

คลาส `SalariedEmployee` สืบทอดจากคลาส `Employee` และโอเวอร์ไรด์เมทอดนามธรรม `getPaymentAmount` ซึ่งรับมาจากอินเทอร์เฟซ `Payable` ผ่านซูเปอร์คลาส `Employee` ด้วย เมทอด `getPaymentAmount` ที่โอเวอร์ไรด์นี้คำนวณยอดที่ต้องชำระจากเงินเดือน

````java
public class SalariedEmployee extends Employee {
    private double salary;
    
    public SalariedEmployee(String firstName, String lastName, String nationalIdNumber, double salary) {
        super(firstName, lastName, nationalIdNumber);
        
        if (salary < 0.0) {
            throw new IllegalArgumentException("Salary must be >= 0.0");
        }
        
        this.salary = salary;
    }

    public void setSalary(double salary) {
        if (salary < 0.0) {
            throw new IllegalArgumentException("Salary must be >= 0.0");
        }
        
        this.salary = salary;        
    }
    
    public double getSalary() {
        return salary;
    }
    
    @Override
    public double getPaymentAmount() {
        return getSalary();
    }

    @Override
    public String toString() {
        return String.format("Salaried Employee: %s%n%s: %,.2f", super.toString(), "Salary", getSalary());
    }
}
````

เราทดสอบโค้ดด้านบนนี้ในคลาส `PayableInterfaceTest` โดยการสร้างอาร์เรย์ของ `Payable` ขึ้นมา แล้วบรรจุอ็อบเจกต์จากทั้ง `Invoice` และ `SalariedEmployee` ลงไป

````java
public class PayableInterfaceTest {
    public static void main(String[] args) {
        Payable[] payableObjects = new Payable[4];
        
        payableObjects[0] = new Invoice("01234", "Seat", 2, 10_000.00);
        payableObjects[1] = new Invoice("56789", "Tire", 4, 2_500.00);
        payableObjects[2] = new SalariedEmployee("John", "Smith", "111-11-1111", 100_000.00);
        payableObjects[3] = new SalariedEmployee("Lisa", "Barnes", "888-88-8888", 150_000.00);
        
        System.out.println("Invoices and Employees processed polymorphically:");
        
        for (Payable currentPayable : payableObjects) {
            System.out.printf("%n%s%nPayment Due: %,.2f%n", currentPayable, currentPayable.getPaymentAmount());
        }
    }
}
````

เห็นได้ว่าเราสามารถใช้ตัวแปรชนิด `Payable` อ้างอิงได้ทั้งอ็อบเจกต์ที่มาจากคลาส `Invoice` และคลาส `SalariedEmployee` โดยที่คลาสทั้งสองนี้มีการอิมพลีเมนต์อินเทอร์เฟซ `Payable` (สำหรับคลาส `SalariedEmployee` เป็นการอิมพลีเมนต์ทางอ้อมผ่านซูเปอร์คลาส `Employee`) แสดงให้เห็นลักษณะการมีหลายรูปแบบผ่านอินเทอร์เฟซ

อินเทอร์เฟซ `Payable` ทำให้เราสามารถเรียกใช้เมทอด `getPaymentAmount` ได้โดยไม่ต้องสนใจว่าอ็อบเจกต์นั้นจริง ๆ แล้วมาจากคลาสใด ตราบเท่าที่คลาสนั้นอิมพลีเมนต์อินเทอร์เฟซ `Payable`

อินเทอร์เฟซมีสัญลักษณ์ในแผนผังคลาส UML ดังภาพด้านล่างนี้ โดยจะมี `<<interface>>` กำกับเพื่อระบุว่าเป็นอินเทอร์เฟซ และคลาสที่อิมพลีเมนต์อินเทอร์เฟซจะแสดงความสัมพันธ์ด้วยเส้นประ

ในไลบรารีมาตรฐานของ Java มีอินเทอร์เฟซที่กำหนดมาให้แล้วอยู่จำนวนหนึ่ง อินเทอร์เฟซมาตรฐานที่มักจะพบการใช้งานบ่อยได้แก่

- `Comparable` คลาสที่ต้องการให้อ็อบเจกต์สามารถเปรียบเทียบลำดับมากกว่าน้อยกว่าได้จะอิมพลีเมนต์อินเทอร์เฟซนี้ เราจะใช้ `Comparable` ในตอนที่เราพูดถึงเรื่องเจเนอริกส์ (generics)
- `Serializable` อิมพลีเมนต์โดยคลาสที่ต้องการให้อ็อบเจกต์สามารถบันทึกลงที่เก็บข้อมูลต่าง ๆ หรือส่งผ่านเครือข่ายได้
- `Runnable` อินเทอร์เฟซนี้ระบุเมทอด `run` เอาไว้ และจะอิมพลีเมนต์โดยคลาสที่แทนงานที่ต้องรัน มักจะใช้กับการทำงานแบบหลายเทรดพร้อมกัน (multithreading)
- กลุ่ม GUI event listener การทำงานกับ GUI จะต้องมีการจัดการกับเหตุการณ์ต่าง ๆ ที่เกิดขึ้นจากการที่ผู้ใช้ติดต่อกับส่วนต่อประสานผู้ใช้ เราจะอิมพลีเมนต์อินเทอร์เฟซในกลุ่มนี้เพื่อสร้างตัวจัดการเหตุการณ์
- `AutoCloseable` อิมพลีเมนต์โดยคลาสที่ตั้งใจให้สามารถนำไปใช้กับคำสั่ง `try`-with-resources ได้
