# การมีหลายรูปแบบ

คำว่า polymorphism เป็นคำประสมจากคำในภาษากรีก _polys_ แปลว่า "มาก" หรือ "หลาย" และ _morphē_ แปลว่า "รูปแบบ" หรือ "รูปร่าง" รวมกันจึงมีความหมายว่า "การมีหลายรูปแบบ" ในที่นี้เราจะใช้คำว่าการมีหลายรูปแบบและคำว่าโพลีมอร์ฟิซึมแทนความหมายเดียวกันตามความเหมาะสมในแต่ละบริบท

โพลีมอร์ฟิซึมที่ปรากฏในการโปรแกรมมีหลายชนิด เช่น

- **โพลีมอร์ฟิซึมแบบเฉพาะกิจ (ad hoc polymorphism)** การที่ฟังก์ชันหรือเมทอดสามารถแสดงพฤติกรรมที่แตกต่างกันโดยขึ้นอยู่กับรูปแบบของอาร์กิวเมนต์ที่ส่งผ่านเข้ามา การมีหลายรูปแบบแบบนี้จะขึ้นอยู่กับลักษณะเฉพาะเจาะจงของชนิดของอาร์กิวเมนต์แต่ละแบบไป จึงเรียกว่าเฉพาะกิจ (ad hoc) มักจะปรากฏในภาษาโปรแกรมในรูปของการโอเวอร์โหลดฟังก์ชันหรือเมทอด ซึ่งเป็นการประกาศฟังก์ชันหรือเมทอดหลายอันในชื่อเดียวกัน แต่มีชนิดของพารามิเตอร์ที่ต่างกัน

- **โพลีมอร์ฟิซึมแบบอิงพารามิเตอร์ (parametric polymorphism)** 

- **การกำหนดชนิดย่อย (subtyping)** การที่เราสามารถใช้ชื่อชื่อเดียวหรือตัวแปรตัวเดียวกันอ้างอิงข้อมูลหรืออ็อบเจกต์ที่ต่างคลาสกัน โดยที่คลาสที่ต่างกันเหล่านั้นมีซูเปอร์คลาสร่วมกัน กล่าวคือ เราสามารถอ้างอิงทั้งอ็อบเจกต์จากซูเปอร์คลาสและจากซับคลาสได้ด้วยตัวแปรตัวเดียวกัน คำว่าชนิดย่อยในที่นี้หมายถึงซับคลาส

- **การกำหนดชนิดแบบเป็ด (duck typing)**

ในบทนี้เราจะพิจารณาเฉพาะการกำหนดชนิดย่อย (subtyping) เนื่องจากในการโปรแกรมแบบ OOP เวลาที่เราพูดถึงโพลีมอร์ฟิซึมเราจะหมายถึงโพลีมอร์ฟิซึมแบบนี้ ภายใต้บริบทนี้ เราสามารถนิยามคำว่าโพลีมอร์ฟิซึมได้ว่า
